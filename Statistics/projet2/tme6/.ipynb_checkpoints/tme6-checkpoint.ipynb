# TME 6- Simulation de microfermes en maraîchage diversifié

Dans les microfermes en maraîchage diversifié, les produits récoltés sont souvent vendus en direct aux particuliers sous forme de paniers hebdomadaires. Il faut pour satisfaire la clientèle avoir suffisament de sortes de légumes et suffisament de sortes de légume pour chaque catégorie. Dans la suite, on considère 8 catégories de légume. Le maraîcher choisit donc parmi les **cycles de culture** ceux qui sont adaptés pour respecter les contraintes de quantité et de diversité. La viabilité d'une microferme respectant ces contraintes sera évaluées sur la base du revenu et de la charge de travail qu'elle génère. Le but du projet est de simuler des microfermes de manière aléatoire et d'évaluer leur viabilité.

Le but de ce TME est de se donner les moyens d'étudier la viabilité de tels microfermes. Pour se faire, il nous faudrait un échantillon de microfermes sur lequel faire nos calculs statistiques (rendements moyens, etc.). Mais nous n'avons pas cet échantillon.

Nous allons donc le créer de notre mieux en utilisant un certain nombres d'informations qui nous permettrons de proposer des microfermes raisonnalbes.

On dispose de deux bases de données issues de la thèse de Kevin Morel qui rassemble les plus de $1000$ cycles de culture et qui donne les critères de quantité et de diversité.

- l'unité de production s'appelle le "cycle". Chaque ferme choisit plusieurs cycles dans **"cropping_cycles.csv"**
- l'unité de vente pour la ferme est le "lot" (**share**). Chaque cycle peut produire un certain nombre de lots en fonction du mois.
- l'unité d'achat pour le client est le "panier" composé d'un certain nombre de lots.
  
- La base de données **"cropping_cycles.csv"** contient plus de $1000$ cycles de culture avec pour chaque cycle de culture:
  - un identifiant entier unique (`ID`), par exemple 233
  - le nom en français du légume concerné (`Crop_french`), par exemple `Tomate cerise`
  - la catégorie du légume concerné (`Crop category`), par exemple `Tomato`
  - le nombre de lots minimum par mois de récolte (`Shmin_month`) et le nombre de lot maximum total (`Shmax_total`) qu'on peut cultiver pour ce cycle
  - la durée de la récolte en mois (`Harvest_total`)
  - pour chaque mois de l'année, un booléen (0 ou 1) qui indique si le légume cultivé dans ce cycle peut être vendu (`Sale_Jan`, `Sale_Feb`, `Sale_Mar`,...)

  Pour une micro-ferme, il s'agit donc de sélectionner les cycles qui vont lui permettre de répondre à la demande commerciale.

- La base de données **"demandes.csv"** qui contient les besoins de légumes dans les paniers hebdomadaires. En particuleir, il décrit pour chaque catégorie de légume:
  - le nom de la categorie (`Crop category`) (par exemple `Tomato`)
  - pour chaque mois, le nombre minimum de légumes différent qui doivent être proposées (`Minimal number of crops_Jan`, `Minimal number of crops_Feb`, ..., `Minimal number of crops_Dec`)
  - pour chaque mois, la quantité minimum en nombre de lots de cette catégoir de légume proposé dans cette catégorie (`Minimal quantity of shares_Jan`, `Minimal quantity of shares_Feb`,..., `Minimal quantity of shares_Dec`)
  
Dans la base de données, plusieurs scénarios sont disponibles, ici on se restreindra à une stratégie de vente sur 12 mois (`Marketing`=12M) et à des climats modérés (`Climate`=`Mild`)

```python
marketing="12M"
climat="Mild"
```

#### 2.1 Écrire la fonction `choose_cycle(cycles,demands,category,month)` qui rend l'`id` du cycle choisi ou `False` si elle a rencontré une impossibilité (pas de demandes ou pas de cycles sélectionnables)

## 4- construction d'une micro-ferme valide

Nous savons donc proposer un agenda pour une microferme qui pourra constituer un panier chaque mois avec ses cycles. Mais ces paniers doivent vérifier certains critères (on voit par exemple que notre ferme ci-dessus n'a pas du tout de panier pour les mois de Mai à Juillet).

Les critères sont décrits dans `demands`: il faut une certaine quantité et une certain diversité :
- **quantité** par mois et par catégorie, `demands`fournit une "Minimal quantity of shares" que doit vérifier la qté produites ce mois et pour cette catégorie
- **diversité** par mois et par catégorie, le nombre de cycles (ou crops)  doit dépasser chaque mois un "Minimal number of crops"


### 4.1- Implémenter la fonction `check_qty(agenda,cycle,demands)` qui renvoie la liste des couples `(mois,catégorie)` qui ne respectent pas cette contrainte

Pour chaque mois, pour chaque catégorie dans `demands`, et si `Minimal quantity of shares>0`, vérifier que la somme des productions de tous les légumes de cette catégorie dans ce mois dépasse ce minimum et retourner la liste des défauts.


```python
print(tme6.check_qty(agenda,cycles2,demands))
```

### 4.2- implémenter la fonction `check_diversity(agenda,cycles,demands)` qui renvoie pour chaque mois, le nombre de mois où il n'y a pas assez de légumes différents par catégorie

Pour chaque mois,  et pour chaque catégorie dans `demands`, vérifier si `Minimal number of crops` est plus petit que le nombre de légumes de cette catégorie et retourner la liste des défauts.

```python
print(tme6.check_diversity(agenda,cycles2,demands))
```
